```python
import pandas as pd
```

```python
df_performances = pd.read_csv("../../results_training_exploring/PHA/explored_performances.csv")
df_selected = pd.read_csv("../../results_training_exploring/PHA/selected_models_for_tuning.csv")

df_merge = df_selected.merge(right=df_performances, left_on=['Algorithm','Encoder'], right_on=["algorithm", "encoder"])
df_merge.columns
```

```python
df_merge = df_merge[['Algorithm', 'Encoder', 'F1_cv', 'recall_cv', 'precision_cv', 'accuracy_cv',
       'accuracy_val', 'precision_val', 'recall_val', 'f1_val', 'matthews_corrcoef_val', 'sensitivity', 'specificity', 'Strategy']]

df_merge.head(5)
```

```python
grouped_data = df_merge.groupby(by=['Algorithm', 'Encoder']).mean(numeric_only=True).sort_values(by="matthews_corrcoef_val", ascending=False).reset_index()
grouped_std = df_merge.groupby(by=['Algorithm', 'Encoder']).std(numeric_only=True).sort_values(by="matthews_corrcoef_val", ascending=False).reset_index()
```

```python
grouped_data = grouped_data[["Algorithm", "Encoder", "accuracy_val", "precision_val", "recall_val", "f1_val", "matthews_corrcoef_val", "sensitivity", "specificity"]]
grouped_std = grouped_std[["Algorithm", "Encoder", "accuracy_val", "precision_val", "recall_val", "f1_val", "matthews_corrcoef_val", "sensitivity", "specificity"]]
grouped_std.columns = ["Algorithm", "Encoder", "std_accuracy_val", "std_precision_val", "std_recall_val", "std_f1_val", "std_matthews_corrcoef_val", "std_sensitivity", "std_specificity"]
```

```python
merge_data = grouped_data.merge(right=grouped_std, on=["Algorithm", "Encoder"])
```
