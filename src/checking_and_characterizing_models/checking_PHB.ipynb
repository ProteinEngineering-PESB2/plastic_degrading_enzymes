```python
import pandas as pd
```

```python
df_performances = pd.read_csv("../../results_training_exploring/PHB/explored_performances.csv")
df_selected = pd.read_csv("../../results_training_exploring/PHB/selected_models_for_tuning.csv")

df_merge = df_selected.merge(right=df_performances, left_on=['Algorithm','Encoder'], right_on=["algorithm", "encoder"])
df_merge.columns
```

```python
df_merge = df_merge[['Algorithm', 'Encoder', 'F1_cv', 'recall_cv', 'precision_cv', 'accuracy_cv',
       'accuracy_val', 'precision_val', 'recall_val', 'f1_val', 'matthews_corrcoef_val', 'sensitivity', 'specificity', 'Strategy']]

df_merge.head(5)
```

```python
df_merge.groupby(by=['Algorithm', 'Encoder']).mean(numeric_only=True).sort_values(by="matthews_corrcoef_val", ascending=False)
```

```python
df_merge.groupby(by=['Algorithm', 'Encoder']).std(numeric_only=True).sort_values(by="matthews_corrcoef_val", ascending=False)
```
