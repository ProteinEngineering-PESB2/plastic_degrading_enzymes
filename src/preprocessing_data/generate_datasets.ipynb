```python
import pandas as pd
import os
```

```python
df_data = pd.read_csv("../../raw_data/enzymes_plastics.csv")
df_data.head()
```

```python
df_data.columns = ['sequence', 'ECOFLEX', 'ECOVIO-FT', 'LDPE', 'NR', 'O-PVA',
       'P(3HB-CO-3MP)', 'P3HP', 'P3HV', 'P4HB', 'PBAT', 'PBS', 'PBS-BLEND',
       'PBSA', 'PBSA-BLEND', 'PBSET', 'PCL', 'PE', 'PEA', 'PEF', 'PEG', 'PES',
       'PET', 'PHA', 'PHB', 'PHBH', 'PHBV', 'PHBVH', 'PHO', 'PHPV',
       'PHTHALATE', 'PHV', 'PLA', 'PMCL', 'PPL', 'PS', 'PVA', 'TP',
       'unknown_plastic', 'PU_PUR', 'NYLON_PA', 'length', 'is_canon']
```

```python
for plastic in ['PET', 'PHB', 'PHA', 'PLA', 'PCL', 'PU_PUR',
       'NYLON_PA', 'PBAT']:
    
    df_filter = df_data[["sequence", plastic]]
    df_filter.columns = ["sequence", "label"]
    command = f"mkdir -p ../../results/generated_dataset/{plastic}"
    os.system(command)

    df_filter.to_csv(f"../../results/generated_dataset/{plastic}/processed_data.csv", index=False)
```
